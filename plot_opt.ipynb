```python
import matplotlib.pyplot as plt
import numpy as np

# Open the file and read its contents into a string
with open('opt_data.txt', 'r') as file:
    data = file.read()

l=100
# Now 'data' contains the contents of 'opt_data.txt' as a string
#print(data)

# Parsing the data
total_energies = []
entropies = []

# Splitting the data into lines and extracting numbers
for line in data.strip().split("\n"):
    parts = line.split(";")
    energy = float(parts[0].split(":")[1].strip())
    entropy = float(parts[1].split(":")[1].strip())
    total_energies.append(energy)
    entropies.append(entropy)

# Plotting the data
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Step')
ax1.set_ylabel('Total Energy', color=color)
ax1.plot(total_energies[:l], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Entropy', color=color)
ax2.plot(entropies[:l], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('Total Energy and Entropy over Optimization Steps')
plt.show()
```
