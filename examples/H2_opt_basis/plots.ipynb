```python
import json
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
folder = 'res'
```

```python
res = open(folder+'/res_FCI.json')
res = json.load(res)
FCI_sto_3Gs = res['sto-3Gs']
FCI_HFpVDZs = res['HFpVDZs']
FCI_KSpVDZs = res['KSpVDZs'] 
FCI_EGNNpVDZs = res['EGNNpVDZs'] 
bond_lengths = res['bond_lengths']
FCI_fci_energies = res['fci_energies']
```

```python
# Sample data
x = range(10)
y = [i**2 for i in x]

# Create main plot
fig, ax1 = plt.subplots()
ax1.plot(bond_lengths, FCI_sto_3Gs,  label = '#basis={}: sto-3g'.format(4))
ax1.plot(bond_lengths, FCI_HFpVDZs,  label = '#basis={}: HF/cc-pVDZ downfold'.format(4))
ax1.plot(bond_lengths, FCI_KSpVDZs,  label = '#basis={}: KS/cc-pVDZ downfold'.format(4))
ax1.plot(bond_lengths, FCI_EGNNpVDZs,  label = '#basis={}: EGNN/cc-pVDZ downfold'.format(4))
ax1.plot(bond_lengths, FCI_fci_energies,  label = '#basis={}: cc-pVDZ'.format(20))

# Create inset plot
ax2 = inset_axes(ax1, width="60%", height="60%", bbox_to_anchor=(-0.02, 0.3, 0.6, 0.6), bbox_transform=ax1.transAxes) # Adjust size and location
ax2.plot(bond_lengths, np.array(FCI_sto_3Gs)-np.array(FCI_fci_energies), label = '#basis=#qubit={}: sto-3g'.format(4))
ax2.plot(bond_lengths, np.array(FCI_HFpVDZs)-np.array(FCI_fci_energies), label = '#basis=#qubit={}: HF/cc-pVDZ downfold'.format(4))
ax2.plot(bond_lengths, np.array(FCI_KSpVDZs)-np.array(FCI_fci_energies), label = '#basis=#qubit={}: KS/cc-pVDZ downfold'.format(4))
ax2.plot(bond_lengths, np.array(FCI_EGNNpVDZs)-np.array(FCI_fci_energies), label = '#basis=#qubit={}: EGNN/cc-pVDZ downfold'.format(4))
ax2.set_ylabel('(Hartree)')
ax2.set_xlabel('(angstrom)')
#plt.legend()
ax2.set_title('basis error')
#ax2.set_xticks([])  # Optionally remove x-ticks
#ax2.set_yticks([])  # Optionally remove y-ticks

# Add labels, legend, and show plot
ax1.set_ylabel('Energy (Hartree)')
ax1.set_xlabel('Bond length (angstrom)')
ax1.set_title('ED results of H2 molecule')
ax1.legend(fontsize='x-small', markerscale=0.4, handlelength=1)
plt.show()
```

```python


# Sample data
categories = ['sto-3g', 'HF downfold', 'KS downfold','EGNN downfold']
sto_3g_opt_err = np.mean(abs(np.array(VQE_sto_3Gs)-np.array(FCI_sto_3Gs)))
sto_3g_basis_err = np.mean(abs(np.array(FCI_sto_3Gs)-np.array(FCI_fci_energies)))
HFpVDZ_opt_err = np.mean(abs(np.array(VQE_HFpVDZs)-np.array(FCI_HFpVDZs)))
HFpVDZ_basis_err = np.mean(abs(np.array(FCI_HFpVDZs)-np.array(FCI_fci_energies)))
KSpVDZ_opt_err = np.mean(abs(np.array(VQE_KSpVDZs)-np.array(FCI_KSpVDZs)))
KSpVDZ_basis_err = np.mean(abs(np.array(FCI_KSpVDZs)-np.array(FCI_fci_energies)))
EGNNpVDZ_opt_err = np.mean(abs(np.array(VQE_EGNNpVDZs)-np.array(FCI_EGNNpVDZs)))
EGNNpVDZ_basis_err = np.mean(abs(np.array(FCI_EGNNpVDZs)-np.array(FCI_fci_energies)))
optimization_errors = [sto_3g_opt_err, HFpVDZ_opt_err, KSpVDZ_opt_err, EGNNpVDZ_opt_err]
basis_errors = [sto_3g_basis_err, HFpVDZ_basis_err, KSpVDZ_basis_err, EGNNpVDZ_basis_err]
total_errors = np.array(optimization_errors) + np.array(basis_errors)

# Create a stacked bar chart
fig, ax = plt.subplots()
bar1 = ax.bar(categories, optimization_errors, label='Optimization Error', color='skyblue')
bar2 = ax.bar(categories, basis_errors, bottom=optimization_errors, label='Basis Error', color='orange')

# Add labels, title, and legend
#ax.set_xlabel('Categories')
ax.set_ylabel('Errors')
ax.set_title('VQE Error Sources for H2 molecule')
ax.legend()

# Display the plot
plt.show()
```
